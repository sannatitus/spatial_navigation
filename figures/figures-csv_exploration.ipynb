```python
from pathlib import Path

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

px = 1 / plt.rcParams["figure.dpi"]  # pixel in inches
plt.rcParams["figure.figsize"] = (900 * px, 600 * px)
```

```python
data_path = Path(r"C:\Users\jai\Downloads\DLC_jwasp_femaleandmale_results_Sanna.xlsx")
data = pd.read_excel(data_path, sheet_name="results")
data.drop(
    columns=[
        "water_open_time",
        "control_open_time",
        "nectar1_open_time",
        "nectar2_open_time",
    ],
    inplace=True,
)
cols = data.columns
```

```python
# If you want to plot row-by-row
for i in range(len(data)):
    datum = data.iloc[i]
    numeric_datum = datum[cols[2:]]
    numeric_datum.dropna(inplace=True)
    ax = sns.barplot(x=np.arange(len(numeric_datum)), y=numeric_datum)
    ax.set_xticklabels(list(numeric_datum.index), rotation=-20)
    ax.set_title(datum.Subject)
    plt.show()
```

```python
# If you want to plot column-by-column
for i, col in enumerate(data.columns[2:]):
    vals = data[col]
    vals.dropna(inplace=True)
    good_indxs = vals.index
    ax = sns.barplot(x=data.Subject[good_indxs], y=vals)
    ax.set_xticklabels(list(data.Subject[good_indxs].unique()), rotation=-20)
    ax.set_title(col)
    plt.show()
```
