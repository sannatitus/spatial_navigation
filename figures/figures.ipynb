```python
#(on SannaDesk) conda activate matplot
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pathlib import Path
```

```python
df_path = Path(r"X:\processed\LondonZoo\Jewel-wasp_Ampulex-compressa\DLC_femaleandmale\%occupation-results.xlsx")
df = pd.read_excel(df_path, sheet_name="Sheet1")
#ensure correct data with print(data.head())
```

```python
#overall average 
oadf = df.iloc[9:18,:]
print(oadf)
sns.set_style("white")
o = sns.barplot(x = "ROI", y = "%occupation", hue="ROI", palette="Paired", hue_order=["water_open", "water_box", "control_open", "control_box", "nectar1_open", "nectar1_box", "nectar2_open", "nectar2_box", "tube"], dodge=False, data = oadf)
o.set_xticklabels(
    o.get_xticklabels(), 
    rotation=45, 
    horizontalalignment='right'
)
o.set(title="Average % ROI occupation for all wasps in all experimental conditions")
plt.legend(loc='right', bbox_to_anchor=(0.9, 0.2, 0.5, 0.5))
plt.show()
fig = o.get_figure()
fig.savefig(r"C:\Users\Sanna\Desktop\Figs\avg.png", bbox_inches='tight')
plt.close()
```

```python
#overall average+random 
ordf = df.iloc[0:18,:]
print(ordf)
os = sns.catplot(
    data=ordf, kind="bar",
    x = "ROI", y = "%occupation", hue="subject", palette = ["#38761d", "#999999", ""],
    order=["water_open", "water_box", 
                            "control_open", "control_box", 
                            "nectar1_open", "nectar1_box", 
                            "nectar2_open", "nectar2_box", 
                            "tube"], 
    hue_order=["average", "random-occupation-probability"], alpha=.6, height=6
)
os.set_xticklabels(rotation=45)
os.set(title="Average % ROI occupation for all wasps in all experimental conditions versus random occupation")
plt.show()
os.savefig(r'C:\Users\Sanna\Desktop\Figs\avg-random.png', bbox_inches='tight')
plt.close()
```
