```python
# Evaluate whether variance is similar between jwasps & crickets
import scipy.stats as stats

# Create list of arrays, one for each group
species_groups = []
for species, group in grouped_species:
    species_groups.append(group['percent_occupation'].values)

# Perform the Bartlett's test on each group
_, p_value = stats.bartlett(*species_groups)

# Check the result of the test
if p_value < 0.05:
    print("The variances are not equal among the species groups (p = {})".format(p_value))
else:
    print("The variances are equal among the species groups (p = {})".format(p_value))
```

```python
# Kruskal-Wallis to compare jewel wasps & crickets
from scipy.stats import kruskal
# Create a list of unique regions in the data
regions = df['region'].unique().tolist()

# Initialize a list to store the Kruskal-Wallis results for each region
kruskal_results = []

# Loop through each region and perform a Kruskal-Wallis test for each one
for region in regions:
    # Filter the data for the current region
    df_region = df[df['region'] == region]
    
    # Split the data into two groups based on species
    species1 = df_region[df_region['species'] == 'jwasp']['percent_occupation']
    species2 = df_region[df_region['species'] == 'cricket']['percent_occupation']

    # Perform a Kruskal-Wallis test on the two groups of data
    kruskal_result = kruskal(species1, species2)
    
    # Store the results in the kruskal_results list
    kruskal_results.append((region, kruskal_result))

# Print the results of the Kruskal-Wallis test for each region
for region, result in kruskal_results:
    print("Kruskal-Wallis Results for Region {}:".format(region))
    print("Statistic: {:.3f}".format(result.statistic))
    print("p-value: {:.3f}".format(result.pvalue))
    print("\n")
```

```python
# Mann-Whitney U test to compare crickets versus probaility of random occupation 
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu

# Define the probability of random occupation for each region
random_occupation = {'shelter': 0.69444,
                     'control1_open': 0.694444444,
                     'control1_box': 1.38889,
                     'control2_open': 0.694444444,
                     'control2_box': 1.38889,
                     'food1_open': 0.694444444,
                     'food1_box': 1.38889,
                     'food2_open': 0.694444444,
                     'food2_box': 1.38889}

# Create a new column in the data frame with the corresponding random occupation
df['random_occupation'] = df['region'].map(random_occupation)

# Group the data by species and region
grouped = df.groupby(['species', 'region'])

# Perform the Mann-Whitney U test for each group
for name, group in grouped:
    species, region = name
    obs = group['percent_occupation'].values
    exp = group['random_occupation'].values[0] * np.ones(len(obs))
    u_stat, p_value = mannwhitneyu(obs, exp)
    print("Species:", species, "Region:", region)
    print("Mann-Whitney U statistic:", u_stat)
    print("p-value:", p_value)
    print("")
```
